# FUNCTIONS

```python
# LOAD PRE TRAINED MODEL
def load_model(file_name = "model.pickle"):
    try:
        device,_ = h.check_device()
        model_dict = torch.load(file_name,map_location=torch.device('cpu'))
        CTX = model_dict['context']
        CTX.device = device
        CTX.prints= False
        W_dict = model_dict['model']
        NET = ModifiedNet(CTX).to(CTX.device)
        NET.load_state_dict(W_dict)
        return NET, CTX
    except Exception as e:
        print(f"Failed to load model with following error:\n{e}")
        return None, None

# SETUP DEFAULT BACKGROUND FOR A PLOT
def ax_style_setup(ax):
    ax.set_facecolor('#eff')
    ax.set_axisbelow(True)
    ax.grid(color='grey', linestyle='dashed')

# COMPUTE THE LINEAR REGRESSION LINE WITH BIAS AND WEIGHT
def linear_reg(x,y):
    x = np.array(x)
    x = np.concatenate( ( x, np.ones(len(x))[:,None] ),axis=1 )
    return np.dot( np.linalg.pinv(x),y)[:,0]

# PLOT FUNCTION FOR GRID SEARCH DATA
def plot_gpgs(search_dict, CTX, results_filename, fig, rotation = [35,35]):
    # SETUP
    ax_left = fig.add_subplot(int('32'+str(len(fig.axes)+1)),projection='3d')
    ax_right = fig.add_subplot(int('32'+str(len(fig.axes)+1)),projection='3d')
    CTX.init_grid(search_dict)
    GPGS = GP_GRID_SEARCH(CTX)
    GPGS.add_data_from_file(const.RESULTS_DIR+results_filename)
    empirical_min,posterior_min,next_pred,epochs=GPGS.predict_minimum()
    print(posterior_min)
    xx,yy = np.meshgrid(CTX.gpgs_parameters['grid'][0],CTX.gpgs_parameters['grid'][1])
    x_min,y_min = GPGS.get_posterior_minimum()
    print(x_min)
    
    # MEAN AND STANDARD DEVIATION
    m = np.reshape(GPGS.mu, (-1,CTX.gpgs_parameters['precision']))
    std = np.reshape(np.sqrt(np.diag(GPGS.Sigma)),(-1,CTX.gpgs_parameters['precision']))
    upper,lower = m+std,m-std

    #### <  LEFT PLOT < ####  
    ax_left.plot_wireframe(xx,yy,m,color='red',alpha=0.5)
    ax_left.plot_surface(xx,yy,m,color='black',alpha=0.1)
    ax_left.scatter(GPGS.DATA_X[:,0],GPGS.DATA_X[:,1],GPGS.DATA_y,linewidths=5,color='black')
    ax_left.scatter(x_min[0],x_min[1],y_min,linewidths=5,s=300,color='green',marker="*")
    ax_left.set_xlabel(CTX.gpgs_parameters['param_labels'][0],fontsize=14)
    ax_left.set_ylabel(CTX.gpgs_parameters['param_labels'][1],fontsize=14)
    ax_left.view_init(rotation[0],rotation[1])
    for i,m_i in enumerate(GPGS.kernel_posterior(Xp=GPGS.DATA_X)[0]):
        ax_left.plot([GPGS.DATA_X[i,0]]*2,[GPGS.DATA_X[i,1]]*2,[m_i,GPGS.DATA_y[i]],color='grey')
    
    #### > RIGHT PLOT > ####
    ax_right.plot_surface(xx,yy,upper,color='grey',alpha=0.6)
    ax_right.plot_surface(xx,yy,m,color='red',alpha=1)
    ax_right.plot_surface(xx,yy,lower,color='grey',alpha=0.6)
    ax_right.set_xlabel(CTX.gpgs_parameters['param_labels'][0],fontsize=14)
    ax_right.set_ylabel(CTX.gpgs_parameters['param_labels'][1],fontsize=14)
```

## KINASE FAMILY

```python
def count_sorted_kinase_families(family_freq,kinases,kinase_families):
    for k in kinases:
        try:
            fam = kinase_families[k]
            family_freq[fam]+=1
        except:
            family_freq['Unknown']+=1
        sorted_families = {k: v for k, v in sorted(family_freq.items(), key=lambda item: item[1],reverse=True)}
    return sorted_families,sum(family_freq.values())

number_of_families = 25
            
# Get kinases
family_freq_train = {f:0 for f in DATA.family_token_keys.keys()}
family_freq_test1 = {f:0 for f in DATA.family_token_keys.keys()}
family_freq_test2 = {f:0 for f in DATA.family_token_keys.keys()}
family_freq_train['Unknown']=0
family_freq_test1['Unknown']=0
family_freq_test2['Unknown']=0

kinases_train = list(DATA.keys_train['target_id'])
kinases_test1 = list(DATA.keys_test[0]['target_id'])
kinases_test2 = list(DATA.keys_test[1]['target_id'])

# Get List of families sorted by frequency
sorted_family_freq_train,N_train = count_sorted_kinase_families(family_freq_train,kinases_train,DATA.kinase_family)
sorted_family_freq_test1,N_test1 = count_sorted_kinase_families(family_freq_test1,kinases_test1,DATA.kinase_family)
sorted_family_freq_test2,N_test2 = count_sorted_kinase_families(family_freq_test2,kinases_test2,DATA.kinase_family)   
plotted_families=['Unknown']

# Plotted Families
for family_dict in [sorted_family_freq_train,sorted_family_freq_test1,sorted_family_freq_test2]:
    for fam in list(family_dict.keys())[:number_of_families]:
        if fam not in plotted_families: plotted_families.append(fam)
plotted_families+=['Other']

sorted_family_freq_train['Other'] = sum(sorted_family_freq_train[k] for k in sorted_family_freq_train.keys() if k not in plotted_families)
sorted_family_freq_test1['Other']  = sum(sorted_family_freq_test1[k] for k in sorted_family_freq_train.keys() if k not in plotted_families)
sorted_family_freq_test2['Other']  = sum(sorted_family_freq_test2[k] for k in sorted_family_freq_train.keys() if k not in plotted_families)
```

```python
plt.style.use('default')
xpos = np.arange(len(plotted_families))
fig, ax = plt.subplots(figsize=(15,10))
ax_style_setup(ax)

width = 0.3

ax.bar(xpos-width, [100*sorted_family_freq_train[f]/N_train for f in plotted_families],width,label='Training Dataset',color='orange')
ax.bar(xpos, [100*sorted_family_freq_test1[f]/N_test1 for f in plotted_families],width,label='Test Dataset 1',color='blue')
ax.bar(xpos+width, [100*sorted_family_freq_test2[f]/N_test2 for f in plotted_families],width,label='Test Dataset 2',color='#BC0000')
ax.set_ylabel('Prosentage of kinases belonging to family',fontsize=20)

ax.set_xlim(-0.7,len(plotted_families)-0.4)
ax.set_xticks(xpos)
ax.set_xticklabels(plotted_families,fontsize=15,rotation='vertical')
ax.legend(fontsize=15,shadow=True,facecolor='black',labelcolor='white')
if save_plots: plt.savefig("FamiliesDist.png")
```

# MODEL PERFORMANCE ANALYSIS

```python
def evaluate_dataset(NET,X,y):
    with torch.no_grad():
        y_pred = NET.forward(X)
        y_pred[y_pred<5]=5
        rmse = torch.sqrt( criterion(y_pred,y) ).item()
        
        y_pred_scipy = [x.item() for x in list(y_pred)]
        y_scipy = [x.item() for x in list(y)]
        
        spearman = stats.spearmanr(y_pred_scipy,y_scipy)[0]
        pearson = stats.pearsonr(y_pred_scipy,y_scipy)[0]
        
        return rmse,spearman,pearson,y_pred,y,y-y_pred

criterion = torch.nn.MSELoss()
X_validation = DATA.validation_batch()
rmse_validation,spearman_validation,pearson_validation,pred_validation,true_validation,error_validation = evaluate_dataset(NET,X_validation[0:4],X_validation[4])
X_test1 = DATA.prepare_test_batch(0)
rmse_test1,spearman_test1,pearson_test1,pred_test1,true_test1,error_test1 = evaluate_dataset(NET,X_test1[0:4],torch.unsqueeze(X_test1[4],1))
X_test2 = DATA.prepare_test_batch(1)
rmse_test2,spearman_test2,pearson_test2,pred_test2,true_test2,error_test2 = evaluate_dataset(NET,X_test2[0:4],torch.unsqueeze(X_test2[4],1))
```

```python
def plot_correlation_plot(y,y_pred,ax,reg_line_edges=np.array([5,10.5])):
    ax.scatter(y, y_pred,marker ="x",c='#205')
    ax.set_aspect('equal','box')
    ax.set_xticks([5,6,7,8,9,10])
    ax.set_ylabel('Predicted value',fontsize=15)
    a,b = linear_reg( y,y_pred )
    ax.plot(reg_line_edges, b+a*reg_line_edges)
```

```python
fig,axes = plt.subplots(nrows=3, figsize=[10,20],sharex=True,sharey=True)
for ax in axes:
    ax_style_setup(ax)
    
# VALIDATION SET
axes[0].set_title('Validation Dataset',fontsize=25)
plot_correlation_plot(true_validation,pred_validation,axes[0])

# TEST SET 1
axes[1].set_title('Test Dataset1',fontsize=25)
plot_correlation_plot(true_test1,pred_test1,axes[1])

# TEST SET 2
axes[2].set_title('Test Dataset2',fontsize=25)
plot_correlation_plot(true_test2,pred_test2,axes[2])
axes[2].set_xlabel('True value',fontsize=15)

plt.tight_layout()
if save_plots: fig.savefig('correlation_plot.png')
```

```python
validation_results = [rmse_validation,spearman_validation,pearson_validation]
test1_results = [rmse_test1,spearman_test1,pearson_test1]
test2_results = [rmse_test2,spearman_test2,pearson_test2]
results = pd.DataFrame({'validation':validation_results,'test1':test1_results,'test2':test2_results},index = ['rmse','spearman','pearson']) 
if save_plots: results.to_csv('final_results.csv')
```
